# There is an imbalance situation in dataset, we need to resample the dataset

```python
from imblearn.over_sampling import SMOTE
```

```python
# Resampling the minority class. The strategy can be changed as required.
sm = SMOTE(sampling_strategy='minority', random_state=42)
# Fit the model to generate the data.
oversampled_X, oversampled_Y = sm.fit_resample(data.drop(' Label', axis=1), data[' Label'])
oversampled = pd.concat([pd.DataFrame(oversampled_Y), pd.DataFrame(oversampled_X)], axis=1)
```

```python
oversampled[' Label'].value_counts()
```

```python
data_X=oversampled.drop([' Label'], axis=1)
data_X.info()
```

```python
data_Y=oversampled[' Label']
```

Now, we do a feature selection process by using Boruta algorithm which is based on random forest algorithm

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
```

```python
rfc=RandomForestClassifier(random_state=1, n_estimators=1000, max_depth=5)
```

```python
boruta_selector=BorutaPy(rfc,n_estimators='auto', verbose=2, random_state=1)
boruta_selector.fit(np.array(data_X), np.array(data_Y))
```

```python
selected_rf_features = pd.DataFrame({'Feature':list(data_X.columns),
                                      'Ranking':boruta_selector.ranking_})
selected_rf_features.sort_values(by='Ranking')
```

```python
data_X=boruta_selector.transform(np.array(data_X))
```

```python
data_X.shape
```

```python
data_Y.replace('BENIGN',0, inplace=True)
```

```python
data_Y.replace('DDoS',1, inplace=True)
```

```python
data_Y.value_counts()
```

```python
data_Y.info()
```

# Feature Normalization

```python
from sklearn import preprocessing
```

```python
min_max_scaler=preprocessing.MinMaxScaler()
data_X=min_max_scaler.fit_transform(data_X)
```

# Split dataset into train set ans test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size=0.2)#split the data
```

# Building model

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
```

```python
#Model ANN
model_ANN = Sequential([
    Dense(100, activation='relu', input_dim=66),
    Dense(60, activation='relu'),
    Dense(40, activation='relu'),
    Dense(1, activation='sigmoid')
])
model_ANN.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
```

```python
history = model_ANN.fit(X_train, Y_train, epochs = 100,validation_split=0.2, verbose = 1, batch_size=50)
```

# test the model

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns 
import matplotlib.pyplot as plt
y_pred=(model_ANN.predict(X_test)>0.8).astype("int32")
conf_mat = confusion_matrix(Y_test, y_pred)
fig, ax = plt.subplots(figsize=(8,8))
sns.heatmap(conf_mat, annot=True, cmap="Blues", fmt='d')
plt.ylabel('Actual')
plt.xlabel('Predicted')
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
accuracy = accuracy_score(Y_test, y_pred)
print('Accuracy: %f' %accuracy)
precision = precision_score(Y_test, y_pred)
print('Precison: %f' %precision)
recall = recall_score(Y_test, y_pred)
print('Recall: %f' %recall)
f1 = f1_score(Y_test, y_pred)
print('F1: %f' %f1)
```
